```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import os
import json
from PIL import Image
import pickle
```

```python
A = np.array(pickle.load(open("D:/tig_welding_dataset/al5083/train_short/A.pickle","rb")))

B= np.array(pickle.load(open("D:/tig_welding_dataset/al5083/train_short/B.pickle","rb")))
```

```python
# Check the shape of the array
image_shape = A[0].shape
print("Image size:", image_shape)
print(len(B))
print(len(A))
```

```python
IMG_SIZE_1=400
IMG_SIZE_2=487
```

```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet169
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Flatten,Reshape
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import TensorBoard
# Load the pre-trained DenseNet-169 model
base_model = DenseNet169(weights='imagenet', include_top=False, input_shape=(IMG_SIZE_1, IMG_SIZE_2, 3))

# Freeze the pre-trained layers
for layer in base_model.layers:
    layer.trainable = False

# Create a new model
model = Sequential()

# Add the DenseNet-169 base model to the new model
model.add(base_model)

# Add a pooling layer to reduce the spatial dimensions
model.add(Flatten())

# Add the LSTM layer
model.add(Reshape((1, -1)))  # Reshape to (1, flattened_features)
model.add(LSTM(64))

# Add fully connected layers
model.add(Dense(256, activation='relu'))
model.add(Dense(64, activation='relu'))

# Add the output layer
model.add(Dense(1, activation='sigmoid'))  # Binary classification, sigmoid activation

# Define the optimizer
optimizer = SGD(learning_rate=0.001)

# Compile the model
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Create a TensorBoard callback
NAME = "TIG_DenseNet169-BINARY-LSTM_model_"

tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))

# Train the model with TensorBoard callback
model.fit(A, B, batch_size=10, epochs=5, validation_split=0.2, callbacks=[tensorboard])
    
# Save the model
model.save(f'DenseNet169-LSTM-BINARY.h5')
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import TensorBoard

# Load the pre-trained VGG16 model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(IMG_SIZE_1, IMG_SIZE_2, 3))

# Freeze the pre-trained layers
for layer in base_model.layers:
    layer.trainable = False

# Create a new model
model = Sequential()

# Add the VGG16 base model to the new model
model.add(base_model)

# Add fully connected layers
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))

# Add the output layer
num_classes=6
model.add(Dense(num_classes, activation='softmax'))

# Define the optimizer
optimizer = SGD(learning_rate=0.01)

# Compile the model
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Create a TensorBoard callback
NAME = "TIG_VGG_SGD"

tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))


# Train the model with TensorBoard callback
model.fit(A, B, batch_size=10, epochs=5,  validation_split=0.2, callbacks=[tensorboard])

# Save the model
model.save('vgg16_model.h5')
```
