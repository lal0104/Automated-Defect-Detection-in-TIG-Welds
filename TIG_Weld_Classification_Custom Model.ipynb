```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import os
import json
from PIL import Image
```

```python
IMG_SIZE_1=250
IMG_SIZE_2=250

# Path to the "train" folder
train_folder = r"D:\tig_welding_dataset\al5083\train"
#train_folder = r"D:\tig_welding_dataset\al5083\train_short"


# List to store the training data
training_data = []
#A=[]
p = 0
#B=[]
# Load the JSON file and get the labels
json_path = os.path.join(train_folder, "train.json")
with open(json_path, "r") as json_file:
    labels = json.load(json_file)

# Iterate over each folder inside the "train" folder
for folder_name in os.listdir(train_folder):
    folder_path = os.path.join(train_folder, folder_name)
    
    # Check if it's a directory
    if os.path.isdir(folder_path):
        # Iterate over each file inside the folder
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)
            
            file_path2 = os.path.basename(folder_name)+"/"+file_name          
            # Check if it's an image file
            if file_name.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                # Open the image and resize it
                img_array = cv2.imread(file_path ,cv2.IMREAD_COLOR)  # convert to array
                new_array = cv2.resize(img_array, (IMG_SIZE_1,IMG_SIZE_2))  # resize to normalize data size
               
                # Get the label for the current file
                label = labels.get(file_path2)
                
                # Append the resized image and its label to the training data
                if label is not None:
                    training_data.append([new_array, label])
                    #A.append(np.array(new_array))
                    #B.append(label)
        p = p+1
        print(p)
# Print the first image and its label in the training data
if training_data:
    print(training_data[0])
else:
    print("Training data is empty.")
```

```python
import random

random.shuffle(training_data)
print(training_data[0])
```

```python
A = []
B = []

for features,label in training_data:
    A.append(np.array(features))
    B.append(label)

A = np.array(A).reshape(-1, IMG_SIZE_1,IMG_SIZE_2, 3)
B = np.array(B)
```

```python
import pickle

pickle_out1 = open("D:/A.pickle","wb")
#pickle_out1 = open("D:/tig_welding_dataset/al5083/train_short/A.pickle","wb")

pickle.dump(A, pickle_out1)
pickle_out1.close()

pickle_out2 = open("D:/B.pickle","wb")
#pickle_out2 = open("D:/tig_welding_dataset/al5083/train_short/B.pickle","wb")
pickle.dump(B, pickle_out2)
pickle_out2.close()
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
import numpy as np
import pickle

NAME = "TIG_Weld_classification_model_1"

#tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))

A=A/255.0

# Define the model architecture
model = Sequential()
model.add(Conv2D(16, (5, 5), activation='relu', input_shape=(IMG_SIZE_1,IMG_SIZE_2, 3)))
model.add(MaxPooling2D(pool_size=(3, 3), strides=3))
model.add(Conv2D(32, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(3, 3), strides=3))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(3, 3), strides=3))
model.add(Conv2D(128, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(3, 3), strides=3))
model.add(Flatten())
model.add(Dense(384, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(6, activation='softmax'))

# Set the learning rate
learning_rate = 0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate, beta_1=0.9, beta_2=0.999)

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Train the model
batch_size = 10
epochs = 5
model.fit(A,B, batch_size=batch_size, epochs=epochs, validation_split=0.2)

# Save the trained model
model.save('TIG_Weld_classification_model_1')
```
