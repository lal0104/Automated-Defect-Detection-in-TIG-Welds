```python
import numpy as np
import os
import cv2
import os
import json
from PIL import Image
```

```python
IMG_SIZE_1=250
IMG_SIZE_2=250

# Path to the "test" folder
test_folder = r"D:\tig_welding_dataset\al5083\test"

# List to store the training data
testing_data = []

p = 0

# Load the JSON file and get the labels
json_path = os.path.join(test_folder, "test.json")
with open(json_path, "r") as json_file:
    labels = json.load(json_file)

# Iterate over each folder inside the "train" folder
for folder_name in os.listdir(test_folder):
    folder_path = os.path.join(test_folder, folder_name)
    
    # Check if it's a directory
    if os.path.isdir(folder_path):
        # Iterate over each file inside the folder
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)
            
            file_path2 = os.path.basename(folder_name)+"/"+file_name          
            # Check if it's an image file
            if file_name.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                # Open the image and resize it
                img_array = cv2.imread(file_path ,cv2.IMREAD_COLOR)  # convert to array
                new_array = cv2.resize(img_array, (IMG_SIZE_1,IMG_SIZE_2))  # resize to normalize data size
               
                # Get the label for the current file
                label = labels.get(file_path2)
                
                # Append the resized image and its label to the training data
                if label is not None:
                    testing_data.append([new_array, label])
        p = p+1
        print(p)
# Print the first image and its label in the training data
if testing_data:
    print(testing_data[0])
else:
    print("Testing data is empty.")
```

```python
import random

random.shuffle(testing_data)
print(testing_data[0])
```

```python
C = []
D = []

for features,label in testing_data:
    C.append(np.array(features))
    D.append(label)

C = np.array(C).reshape(-1, IMG_SIZE_1,IMG_SIZE_2, 3)
D = np.array(D)
```

```python
import pickle

pickle_out1 = open("D:/C.pickle","wb")

pickle.dump(C, pickle_out1)
pickle_out1.close()

pickle_out2 = open("D:/D.pickle","wb")
pickle.dump(D, pickle_out2)
pickle_out2.close()
```
